```python
# Columns we don't care about

drop_cols = ['RegionCode', 'ConfirmedCases', 'ConfirmedDeaths', 'StringencyIndex',
    'StringencyIndexForDisplay', 'StringencyLegacyIndex',
    'StringencyLegacyIndexForDisplay', 'GovernmentResponseIndex',
    'GovernmentResponseIndexForDisplay', 'ContainmentHealthIndex',
    'ContainmentHealthIndexForDisplay', 'EconomicSupportIndex',
    'EconomicSupportIndexForDisplay', 
    'V1_Vaccine Prioritisation (summary)', 'V1_Notes', 'V2A_Vaccine Availability (summary)', 
    'V2_Notes', 'V2B_Vaccine age eligibility/availability age floor (general population summary)',
    'V2C_Vaccine age eligibility/availability age floor (at risk summary)', 'V2D_Medically/ clinically vulnerable (Non-elderly)',
    'V2E_Education', 'V2F_Frontline workers  (non healthcare)', 'V2G_Frontline workers  (healthcare)' ,
    'V3_Vaccine Financial Support (summary)','V3_Notes' ]

# Columns that identify each measure
id_cols = ['CountryName', 'RegionName', 'Date', 'CountryCode', 'Jurisdiction']

# Filter data for Aruba (so the example is quick. This obv needs to be removed)
# & drop columns we don't care about
# data = ox[ox["CountryName"] == "United States"].drop(columns = drop_cols)
data = ox.drop(columns = drop_cols)

# Get all the stub names using our regular expression
stubs = [re.findall(stub_exp, x) for x in data.columns]

# Flatten the returned list and get only unique stubnames
stubs = list(set([item for sublist in stubs for item in sublist]))    

```

```python
stubs
```

```python
# Function to extract id columns and measure columns
def extract_stub_groups(data: pd.DataFrame, stub: str):

        # Subset ID columns
        id_data = data[id_cols]

        # Subset columns with measure data
        value_data = data.filter(regex=stub)

        # If the measure data has 2 columns, it doesn't have a flag
        if len(value_data.columns) < 3:
            value_data.columns = ["measure", "notes"]
        else:
            value_data.columns = ["measure", "flag", "notes"]

        # cbind these datasets
        combined = pd.concat([id_data, value_data], axis=1)

        # Say what the variable is
        combined["variable"] = stub

        # Set 0s to NA
        combined["measure"] = combined["measure"].replace(0, np.NaN)

        # Drop all nas and return data
        return combined.dropna(subset = ["measure"])

# Do this for one stub (as an example)
# extract_stub_groups(data, stubs[5])

# Do this for all stubs & combine the results
# output_Aruba = pd.concat([extract_stub_groups(data, x) for x in stubs], axis = 0)

# output_Aruba.to_csv('Oxford_separated.csv')


ox_sep = pd.concat([extract_stub_groups(data, x) for x in stubs], axis = 0)


```

```python
def pivot_ox(df1, df2):

    k = pd.pivot_table(df1, index=['CountryName', 'Date', 'RegionName'], columns=[
                      'variable'], values=['Flag', 'measure', 'Notes'], aggfunc=lambda x: ''.join(str(v) for v in x) )

    k.columns = ['_'.join(col[::-1]) if 'measure' not in col else col[-1] for col in k.columns]

    for x, y in prov_dict.items():
        k.rename(columns=prov_dict, inplace=True)

    k.columns

    k = k[sorted(k.columns)]

    df2 = df2.set_index(['CountryName', 'Date', 'RegionName'])

    df2.update(k)

    return df2
```

```python
# Regular Expression for detecting stub names
stub_exp = r'[A-Z][0-9]+'

# URL to get OxCGRT data
url = "https://github.com/OxCGRT/covid-policy-tracker/raw/master/data/OxCGRT_latest_withnotes.csv"

# Read OxCGRT data
# ox = pd.read_csv(url)
ox = pd.read_csv('Not_na.csv')




```

```python
# Columns we don't care about. Note that the Vaccination columns do not need to be included, 
# as they have already been removed in earlier stages.

drop_cols = ['RegionCode', 'ConfirmedCases', 'ConfirmedDeaths', 'StringencyIndex',
     'StringencyIndexForDisplay', 'StringencyLegacyIndex',
     'StringencyLegacyIndexForDisplay', 'GovernmentResponseIndex',
     'GovernmentResponseIndexForDisplay', 'ContainmentHealthIndex',
     'ContainmentHealthIndexForDisplay', 'EconomicSupportIndex',
     'EconomicSupportIndexForDisplay']

# Columns that identify each measure
id_cols = ['CountryName', 'RegionName', 'Date', 'CountryCode', 'Jurisdiction']

# Filter data for Aruba (so the example is quick. This obv needs to be removed)
# & drop columns we don't care about
# data = ox[ox["CountryName"] == "United States"].drop(columns = drop_cols)
data = ox.drop(columns = drop_cols)

# Get all the stub names using our regular expression
stubs = [re.findall(stub_exp, x) for x in data.columns]

# Flatten the returned list and get only unique stubnames
stubs = list(set([item for sublist in stubs for item in sublist]))    
```

```python
# Function to extract id columns and measure columns
def extract_stub_groups(data: pd.DataFrame, stub: str):

        # Subset ID columns
        id_data = data[id_cols]

        # Subset columns with measure data
        value_data = data.filter(regex=stub)

        # If the measure data has 2 columns, it doesn't have a flag
        if len(value_data.columns) < 3:
            value_data.columns = ["measure", "notes"]
        else:
            value_data.columns = ["measure", "flag", "notes"]

        # cbind these datasets
        combined = pd.concat([id_data, value_data], axis=1)

        # Say what the variable is
        combined["variable"] = stub

        # Set 0s to NA
        combined["measure"] = combined["measure"].replace(0, np.NaN)

        # Drop all nas and return data
        return combined.dropna(subset = ["measure"])

# Do this for one stub (as an example)
# extract_stub_groups(data, stubs[5])

# Do this for all stubs & combine the results
# output_Aruba = pd.concat([extract_stub_groups(data, x) for x in stubs], axis = 0)

# output_Aruba.to_csv('Oxford_separated.csv')


ox_sep = pd.concat([extract_stub_groups(data, x) for x in stubs], axis = 0)

output = ox_sep.dropna( how='all',
                    subset=['notes'])

```
